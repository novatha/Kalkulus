# Membuat grafik fungsi dengan bantuan python 3

```python
%matplotlib inline
from scipy import sin, cos, tan, linspace, pi
from matplotlib import *
import matplotlib.pyplot as plt
def f(x) :
    return (x)# dapat diganti sesuai kebutuhan misalnya cos(x) z**x
def derf(f,x):
    # derivative of f(x), dihitung secara numerik
    s=1e-12
    result=(f(x+s)-f(x))/s
    return result
```

```python
x=linspace(-2*pi,2*pi,100)
```

## Plot 

```python
plt.plot(x,f(x),x,derf(f,x))
plt.grid('on')
```
