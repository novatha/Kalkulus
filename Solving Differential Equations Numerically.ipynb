```julia
using DifferentialEquations
```

## Problem Definition
$\frac{du}{dt} = f(u,p,t)$

```julia
k=0.1
f(u,p,t)=k*u
u0=1/2
tspan=(0.0,10.0)
prob=ODEProblem(f,u0,tspan)
```

## Solving the problem

```julia
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)
```

## Plotting

```julia
using Plots
plot(sol,linewidth=5,title="Solution to the linear ODE with a thick line",
     xaxis="Time (t)",yaxis="u(t) (in mikrometer )",label="My Thick Line!") # legend=false
```
